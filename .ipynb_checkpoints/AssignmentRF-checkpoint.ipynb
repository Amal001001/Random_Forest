### split and preprocess the data 

```python
y= bank_df['y'].map({'yes': 1, 'no': 0})
X = bank_df.drop(columns = 'y')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2 ,random_state=3)
```

```python
#get numrical and catgorical varaibles and normalize
cat_features = X_train.describe(exclude =[np.number]).columns.values
numric_features = X_train.describe().columns.values

cat_features
```

```python
# Pipeline
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
# Models
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python

# Create a transformer for numeric columns

numeric_transformer2 = Pipeline(
    steps=[
       
        
        ('scaler', StandardScaler())
    ]
)

# Create Transformer for categorical data

categorical_transformer2 = Pipeline(
    steps=[
        # most_frequent --> mode
      
        ('one_hot', OneHotEncoder(handle_unknown='ignore')) # Ignore unseen categorical in transform step not seen in fit_transform
    ]
)

# Create a preprocessor transformer
preprocessor2 = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer2, numric_features),
        ('cat', categorical_transformer2, cat_features)
    ]
)

# Append classifier to preprocessing pipeline.
# Now we have a full prediction pipeline.
clf2 = Pipeline(
    steps=[
        ('preprocessor', preprocessor2),
        ('classifier', RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0))
    ]
)

```

```python
clf2.fit(X_train, y_train)
```

```python
clf2.score(X_test, y_test)
```

```python
param_grid = {
    "classifier__n_estimators": [5,10,20,30,40], # how many trees in our forest
    "classifier__criterion": ["gini", "entropy"], # what criteria is used to split nodes
    "classifier__max_depth": [2,4,6,8,10] # how deep each decision tree can be
    
}

grid = GridSearchCV(
    clf2,
    param_grid,
    cv = 10,
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train, y_train)
```

```python
grid.score(X_test, y_test)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
print(grid.best_estimator_.named_steps["classifier"].feature_importances_)
```

```python
# Plot Feature Importances to Visualize better

pd.DataFrame(dict(zip(X_train.columns, grid.best_estimator_.named_steps["classifier"].feature_importances_)), index = [1])\
.T\
.sort_values(1, ascending=True)\
.plot(kind="barh", title="Feature Importances (%)", figsize=(10,10));
```

```python
# We can also make a simple confusion matrix
plot_confusion_matrix(grid, X_test, y_test);
```

```python
y_test.value_counts()
```

```python
y_pred = grid.predict(X_test)
```

```python
print(classification_report(y_test, y_pred, target_names=['no','yes']))
```

```python
plt.figure(figsize=(30,30))
tree.plot_tree(
grid.best_estimator_.named_steps["classifier"].estimators_[0], 
    filled=True,
    rounded=True,
    )
plt.show()
```

```python
y= pullution_df_new['CO AQI']
X = pullution_df_new.drop(columns = 'CO AQI')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2 ,random_state=3)
```

```python
#get numrical and catgorical varaibles and normalize
#cat_features = X_train.describe(exclude =[np.number]).columns.values
numric_features = X_train.describe().columns.values

numric_features
```

```python
# Create a transformer for numeric columns

numeric_transformer2 = Pipeline(
    steps=[
       
        
        ('scaler', StandardScaler())
    ]
)

# Create Transformer for categorical data

categorical_transformer2 = Pipeline(
    steps=[
        # most_frequent --> mode
      
        ('one_hot', OneHotEncoder(handle_unknown='ignore')) # Ignore unseen categorical in transform step not seen in fit_transform
    ]
)

# Create a preprocessor transformer
preprocessor2 = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer2, numric_features)
       
    ]
)

# Append classifier to preprocessing pipeline.
# Now we have a full prediction pipeline.
regpip = Pipeline(
    steps=[
        ('preprocessor', preprocessor2),
        ('regresser', RandomForestRegressor(n_estimators = 10, random_state = 0, criterion = 'mse'))
    ]
)
```

```python
regpip.fit(X_train,y_train)
```

```python
regpip.score(X_test, y_test)
```

```python
preds_regpip = regpip.predict(X_test)

mean_absolute_error(y_true=y_test, y_pred=preds_regpip)
```

```python
from sklearn.metrics import r2_score
r2_score(y_true=y_test, y_pred=preds_regpip)
```
